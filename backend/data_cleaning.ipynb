```python
import pandas as pd

raw_transactions = pd.read_csv("../raw_transactions.csv")

display(raw_transactions.tail(50))
```

```python
raw_transactions = raw_transactions.dropna()
raw_transactions["Description"] = raw_transactions["Description"].str.lower()
display(raw_transactions)
```

```python
raw_transactions["Description"] = raw_transactions["Description"].str.replace(r"[^a-zA-Z0-9 ]", " ", regex=True)
raw_transactions["Description"] = raw_transactions["Description"].str.replace(r"\s+", " ", regex=True)
display(raw_transactions.head())
```

```python
raw_transactions["Description"] = raw_transactions["Description"].str.replace(r"[0-9]", " ", regex=True)
raw_transactions["Description"] = raw_transactions["Description"].str.replace(r"\s+", " ", regex=True)
raw_transactions["Description"] = raw_transactions["Description"].str.strip()

display(raw_transactions.head())
```

```python
raw_transactions.to_csv("clean_transactions.csv", index=False)
```
